# <span style="color: #32a83e; border-bottom: 6px solid #8f6621; background-color: #fadb8c; padding-left: 5px; padding-right: 5px;">🍑 10 - Transorm the data with logarithm</span>

![](https://qph.cf2.quoracdn.net/main-qimg-3b2bd9424253097da6d5404ccbf99586-lq)

```python
# Split the train data into X and y
X = train.drop(['smoking'], axis=1)
y = train.smoking

for column in X.columns.tolist():
    X[column] = X[column].apply(lambda x: np.log(x + 1))
    
# Transform test data
for column in test.columns.tolist():
    test[column] = test[column].apply(lambda x: np.log(x + 1))
    
X.hist(color='SkyBlue');
```
